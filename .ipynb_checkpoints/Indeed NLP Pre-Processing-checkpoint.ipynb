## Data Pre-Processing

For this analysis we are going to be primarly analyzing the "rating_descriptions".
Therefore, let's converting the rating_description field to a string format.

```python
df['rating_description'] = df['rating_description'].astype(str)
```

Let's begin by expanding any contractions we might have (ie. "I've" or "I'll").  Keep in mind that this will effectively tokenize the rating descriptions but each contraction wil be a single token.  I other words, "I've" will be "I have" instead of "I", "Have".

```python
df['no_contract'] = df['rating_description'].apply(lambda x: [contractions.fix(word) for word in x.split()])
```

```python
df
```

Now that the contractions are expanded let's turn the list back into a string so we can properly tokenize the expanded contractions as two separate tokens instead of one. 

```python
 df['rating_description_list'] = [' '.join(map(str, l)) for l in df['no_contract']]
```

Now let's tokenize as normal and the expanded contracts will be tokenized accurately.  Since we have some NA values in the rating description column

```python
df['tokenized'] = df['rating_description_list'].apply(word_tokenize)
```

```python
def get_wordnet_pos(tag):
    if tag.startswith('J'):
        return wordnet.ADJ
    elif tag.startswith('V'):
        return wordnet.VERB
    elif tag.startswith('N'):
        return wordnet.NOUN
    elif tag.startswith('R'):
        return wordnet.ADV
    else:
        return wordnet.NOUN
```

```python
df['wordnet_pos'] = df['pos_tags'].apply(lambda x: [(word, get_wordnet_pos(pos_tag)) for (word, pos_tag) in x])
```

## Topic Modeling
Now that we have explored our data and have a good idea behind the relationships between ratings, rating descriptions let's examine the themes or "topics" of our rating descriptions.

We have thoroughly processed our data/text but it is still readable to us humans whereas computers only understand numbers.  Therefore, we need to convert our text into numbers/vectors.  

### CountVectorizer
This method transposes all the words/tokens in a corpus into features/columns, then for each counts the occurence fo each words.  

```python
# you have to feed strings into a countvec
# max_df=0.9 we will remove words which appear in more than 90% of the reviews
# min_df=25 we will remove words which appear in less than 25 reviews

# Creating the vectorizer object
tf_vectorizer = CountVectorizer(max_df=0.9, min_df=25, max_features=1000)

# apply the transformation to create the document term matrix
tf = tf_vectorizer.fit_transform(df['list_lemma'].values.astype('U'))
tf_feature_names = tf_vectorizer.get_feature_names()
```

```python
# looks like we have 441 words in our document term matrix
tf
```

### Latent Dirichlet Allocation (LDA) Approach to Topic Modeling

```python
lda_model = LatentDirichletAllocation(n_components=20, learning_method='online', max_iter=100, random_state=0).fit(tf)
```

```python
print(lda_model)
```

```python
no_top_words = 10

def display_topics(model, feature_names, no_top_words):
    for topic_idx, topic in enumerate(model.components_):
        print("Topic %d:" % (topic_idx))
        print(" ".join([feature_names[i]
                          for i in topic.argsort()[:-no_top_words - 1:-1]]))
              
display_topics(lda_model, tf_feature_names, no_top_words)              
```

```python
# Log Likelyhood: Higher the better
print("Log Likelihood: ", lda_model.score(tf))

# Perplexity: Lower the better. Perplexity = exp(-1. * log-likelihood per word)
print("Perplexity: ", lda_model.perplexity(tf))
```

```python
lda_topic_values = lda_model.transform(tf)
df['lda_topics'] = lda_topic_values.argmax(axis=1) #argmax() returns the position (ie. column name) of the largest value in the row. 
```

### TF-IDF Approach to Text Vectorization
The purpose of using tf-idf is to bring down the weight/impact of tokens which appear very frequently in your corpus|

```python
# Creating our document term matrix
tfidf_vectorizer = TfidfVectorizer(max_df=0.95, min_df = 2, max_features=5000, use_idf=True)
tfidf = tfidf_vectorizer.fit_transform(df['list_lemma'])
tfidf_feature_names = tfidf_vectorizer.get_feature_names()
```

### Non-negative Matrix Factorization (NMF) Approach to Topic Modeling

```python
# LDA cannot use a term matrix derived from tf-idf because it only uses term counts
# NMF can use tf-idf derived term matrix
nmf = NMF(n_components=20, random_state=0, alpha=.1, l1_ratio=.5, init='nndsvd').fit(tfidf)
```

```python
display_topics(nmf, tfidf_feature_names, no_top_words) 
```

```python
nmf_topic_values = nmf.transform(tfidf)
df['nmf_topics'] = nmf_topic_values.argmax(axis=1)
```
